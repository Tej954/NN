```python
#Import required libraries
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Function and its derivative
def f(x):
    return x ** 2
def df(x):
    return 2 * x

# Gradient descent function with tracking of x values
def gradient_descent(starting_point, learning_rate, n_iterations):
    x = starting_point
    x_values = [x]
    for _ in range(n_iterations):
        grad = df(x)
        x = x - learning_rate * grad
        x_values.append(x)
    return x, x_values

# Parameters
starting_point = 10
learning_rate = 0.1
n_iterations = 100
```

```python

# Perform gradient descent
minimum, x_values = gradient_descent(starting_point, learning_rate, n_iterations)

# Create an array of x values for plotting the function
x_plot = np.linspace(-starting_point, starting_point, 400)
y_plot = f(x_plot)

```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='f(x) = x^2')
plt.scatter(x_values, f(np.array(x_values)), color='red', marker='o', label='Gradient Descent Steps')
plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Change the Parameters to see how the minima is achieved
starting_point = 10
learning_rate = 0.9
n_iterations = 100

# Perform gradient descent
minimum, x_values = gradient_descent(starting_point, learning_rate, n_iterations)

# Create an array of x values for plotting the function
x_plot = np.linspace(-starting_point, starting_point, 400)
y_plot = f(x_plot)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='f(x) = x^2')
plt.scatter(x_values, f(np.array(x_values)), color='red', marker='o', label='Gradient Descent Steps')
plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
