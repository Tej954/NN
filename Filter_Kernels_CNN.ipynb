```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('dog1.jpg', cv2.IMREAD_GRAYSCALE)  # Load as grayscale
#print(image.shape)

cv2_imshow(cv2.resize(image, (400, 400)))
cv2.waitKey(0) #waits for these milliseconds before image window is detroyed, if 0 is given then it waits for keypress
```

```python

blur_kernel = np.ones((5, 5), np.float32) / 25


sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]])


# Sobel kernel (horizontal)
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]])

# Sobel kernel (vertical)
sobel_y = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]])


blurred_image = cv2.filter2D(image, -1, blur_kernel)
sharpened_image = cv2.filter2D(image, -1, sharpen_kernel)
edges_x = cv2.filter2D(image, -1, sobel_x)
edges_y = cv2.filter2D(image, -1, sobel_y)

print("\n Blurred Image \n")
cv2_imshow(cv2.resize(blurred_image, (400, 400)))
print("\n Sharpened Image \n")
cv2_imshow(cv2.resize(sharpened_image,(400,400)))
print("\n Edges X \n")
cv2_imshow(cv2.resize(edges_x,(400,400)))
print("\n Edges Y\n")
cv2_imshow(cv2.resize(edges_y,(400,400)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
